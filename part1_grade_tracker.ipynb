```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]


for subject, mark in zip(subjects, marks):
    if 90 <= mark <= 100:
        grade = "A+"
    elif 80 <= mark <= 89:
        grade = "A"
    elif 70 <= mark <= 79:
        grade = "B"
    elif 60 <= mark <= 69:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{subject}: {mark} | Grade {grade}")
```

```python
#total
total = sum(marks)

#average
avg = round(total/len(marks),2)

#Highest
max_marks = max(marks)
max_index = marks.index(max_marks)
highest_subject = subjects[max_index]

#Lowest
min_marks = min(marks)
min_index = marks.index(min_marks)
lowest_subject = subjects[min_index]

# Print results
print(f"Total: {total}")
print(f"Average: {avg}")
print(f"Highest: {highest_subject} ({max_marks})")
print(f"Lowest: {lowest_subject} ({min_marks})")
```

```python
student_name = "Ayesha Sharma"
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]
new_subjects_added = 0

while True:
    subject = input("Enter subject name (or type 'done' to finish): ").strip()
    
    if subject.lower() == "done":
        break
    
    mark_input = input(f"Enter marks for {subject} (0–100): ").strip()
    
    # Validate marks
    try:
        mark = float(mark_input)
        
        if 0 <= mark <= 100:
            subjects.append(subject)
            marks.append(mark)
            new_subjects_added += 1
        else:
            print("⚠️ Invalid marks! Must be between 0 and 100.")
    
    except ValueError:
        print("⚠️ Invalid input! Please enter a numeric value.")

# Final calculations
total = sum(marks)
average = round(total / len(marks), 2)

# Output
print("\n--- Summary ---")
print(f"New subjects added: {new_subjects_added}")
print(f"Updated Average Marks: {average}")
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]


for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    
    print(f"{name} → Average: {avg} | Result: {status}")
```
